Descriptive Statistics with NumPy

```python
temps = df['Temp_C'].to_numpy()
humidity = df['Rel Hum_%'].to_numpy()
wind = df['Wind Speed_km/h'].to_numpy()

print("\n  DESCRIPTIVE STATISTICS ")
print("Mean Temperature:", np.mean(temps))
print("Median Temperature:", np.median(temps))
print("Std Dev Temperature:", np.std(temps))

print("Mean Humidity:", np.mean(humidity))
print("Median Humidity:", np.median(humidity))

print("Mean Wind Speed:", np.mean(wind))
print("Median Wind Speed:", np.median(wind))
```

Monthly & Seasonal Analysis

```python
df['Month'] = df['Date/Time'].dt.month

# Monthly average temperature
monthly_avg = df.groupby('Month')['Temp_C'].mean()
print("\n MONTHLY AVG TEMPERATURE ")
print(monthly_avg)

def season(m):
    if m in [12, 1, 2]: return 'Winter'
    if m in [3, 4, 5]: return 'Spring'
    if m in [6, 7, 8]: return 'Summer'
    return 'Autumn'

df['Season'] = df['Month'].apply(season)
seasonal_avg = df.groupby('Season')['Temp_C'].mean()
print("\n SEASONAL AVG TEMPERATURE ")
print(seasonal_avg)
```

Visualizations with Matplotlib

```python
monthly_avg.plot(kind='line', marker='o')
plt.title("Monthly Average Temperature")
plt.xlabel("Month")
plt.ylabel("Temperature (C)")
plt.show()
```

Visualizations with Barplot

```python
df.groupby('Month')['Wind Speed_km/h'].mean().plot(kind='bar')
plt.title("Average Wind Speed by Month")
plt.xlabel("Month")
plt.ylabel("Wind Speed (km/h)")
plt.show()
```

Histogram  Temperature Distribution

```python
plt.hist(df['Temp_C'])
plt.title("Temperature Distribution")
plt.xlabel("Temperature (C)")
plt.ylabel("Frequency")
plt.show()
```

Identify Extreme Weather

```python
# Days with extremely high temperature (> 30C)
extreme_hot = df[df['Temp_C'] > 30]
print("\n EXTREME HOT DAYS ")
print(extreme_hot[['Date/Time','Temp_C']].head())

# Days with very low visibility (Fog)
extreme_fog = df[df['Visibility_km'] < 2]
print("\n EXTREME FOG DAYS ")
print(extreme_fog[['Date/Time','Visibility_km']].head())
```
